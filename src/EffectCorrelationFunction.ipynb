## Tiling assumptions

The tiling on the sky is done using a chicken wire pattern. To determine the spacing for the underlying triangular mesh we fix the number of pointings and the area to be covered

```python
#n_pointings = 10000
#survey_area = 14000.0 #square degrees

n_pointings  = 100
survey_area = 140.0
focal_plane_area = 8.0
density_centers = n_pointings/survey_area
tile_spacing  = sqrt((4.0/(3.0*sqrt(3.0))) * 1.0/density_centers)
print "Density centers", density_centers
print "Spacing between edges", tile_spacing
```

```python
def generate_tiling_edges(n_x=10, n_y=10, s=1.0):
    """
    Generates the centers for a chicken wire pattern.
    Inputs:
        n_x: number of edges in a triangular pattern along the x_direction
        n_y: number of edges in a triangular pattern along the y_direction
        s: spacing between the edges in the triangular pattern
    Outputs:
        x_edges, y_edges: arrays with the centers in the units provided by the spacing input s.
    """
    x_edges = empty([0])
    y_edges = empty([0])

    if(n_x%2!=0): #just to have a pair number of n_x
        n_x = n_x+1

    template_x = arange(n_x)
    leave = where((template_x%3)!=0)
    template_x = template_x[leave] * s

    for j in range(n_y):
        for center_x in template_x:
            y_edges = append(y_edges, j*s*sqrt(3.0)/2.0)
            if((j%2)==0):
                x_edges = append(x_edges, center_x)
            else:
                x_edges = append(x_edges, center_x-s*1.5)
    return x_edges, y_edges
```

## A simple sky

The galaxies will be distributed following a flat distribution over the sky; the tiling will be done with the chicken wire pattern and the fiber allocation will follow simple heuristics for priorization and final positioning

```python
sky_size_x = sqrt(survey_area)
sky_size_y = sqrt(survey_area)
n_x = int(sky_size_x/tile_spacing)
n_y = int(sky_size_y/(tile_spacing*sqrt(3.0)/2.0))
print sky_size_x, sky_size_y, n_x, n_y
x_pointing, y_pointing = generate_tiling_edges(n_x=n_x,n_y=n_y, s=tile_spacing)
scatter(x_pointing, y_pointing, s =2.0)
print "sky lenght size", sky_size_x
```

```python
def find_galaxy(galaxy_pos_x, galaxy_pos_y, galaxy_ID, galaxy_n_observed, center_x=0.0, center_y=0.0, radius=0.1):
    #find the galaxies that fall inside the observing radius
    index = where((((galaxy_pos_x- center_x)**2 + (galaxy_pos_y - center_y)**2)<radius**2))
    selected_ID=0
    if (size(index) > 0):
        index = index[0]
        if (size(index)==1) :
            selected_ID=index
        else:
            selected_ID=index[0]
    else:
        selected_ID = -1
    return selected_ID
```
