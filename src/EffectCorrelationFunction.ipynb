# Effects of tiling fiber allocation on the correlation function

In this notebook we want to address the following question. What is the effect of assuming a tiling and fiber allocation algorithms to observe a Poissonian galaxy distribution on a flat sky?

In this setup we assume that there are four different kinds of galaxies that can be observed (in orderr of decreasing  priority):

1. Lyman-alpha forest QSOs (QSO-IIs, $z>2.5$)
2. Luminous red galalxies (LRGs).
3. Lyman-alpha forest QSOs (QSO-Is, $z<2.5$).
4. Emission Line Galaxies (ELG)

The number of observations for each one of these galaxy tipes are

1. QSO-IIs require five observations
2. LRGs require two observations.
3. QSO-I require only one observation
4. ELGs have the lowes priority and have at most one observation.

## Tiling assumptions

The tiling on the sky is done using a chicken wire pattern. To determine the spacing for the underlying triangular mesh we fix the number of pointings and the area to be covered

```python
#n_pointings = 10000
#survey_area = 14000.0 #square degrees

n_pointings  = 100.0
survey_area = 140.0


focal_plane_area = 8.0
density_centers = n_pointings/survey_area
tile_spacing  = sqrt((4.0/(3.0*sqrt(3.0))) * 1.0/density_centers)
print "Density centers", density_centers
print "Spacing between edges", tile_spacing
```

```python
def generate_tiling_edges(n_x=10, n_y=10, s=1.0):
    """
    Generates the centers for a chicken wire pattern.
    Inputs:
        n_x: number of edges in a triangular pattern along the x_direction
        n_y: number of edges in a triangular pattern along the y_direction
        s: spacing between the edges in the triangular pattern
    Outputs:
        x_edges, y_edges: arrays with the centers in the units provided by the spacing input s.
    """
    x_edges = empty([0])
    y_edges = empty([0])

    if(n_x%2!=0): #just to have a pair number of n_x
        n_x = n_x+1

    template_x = arange(n_x)
    leave = where((template_x%3)!=0)
    template_x = template_x[leave] * s
    
    template_x_odd = arange(n_x+1) 
    leave = where((template_x_odd%3)!=0)
    template_x_odd = template_x_odd[leave] * s -1.5*s

    for j in range(n_y):
            if((j%2)==0):
                for center_x in template_x:
                    if (center_x >0.0):
                        x_edges = append(x_edges, center_x)
                        y_edges = append(y_edges, j*s*sqrt(3.0)/2.0)
            else:
                for center_x_odd in template_x_odd:
                    if center_x_odd >0.0:
                    
                        x_edges = append(x_edges, center_x_odd)
                        y_edges = append(y_edges, j*s*sqrt(3.0)/2.0)
            
    return x_edges, y_edges
```

# Fiber prioritization heuristics

Here I am following Bob Cahn's document on fiber prioritization

1. If there is a QSO II, assign the fiber to it. In fact, we take the first QSO II if there is more than one. We should also not go beyond 5 observations of the QSO II.

2. If the fiber is not yet assigned and there is a QSO I with no exposures, assign the fiber to it, again taking the first such QSO I.

3. If the fiber is not yet assigned and there is an LRG that already has one exposure, assign the fiber to it. Again, we take the first such LRG.

4. If the fiber is not yet assigned and there is an LRG with no exposures, assign the fiber to it, again taking the first such LRG.

5. If the fiber is not yet assigned and there is an ELG with no exposures, assign the fiber to it.


```python
def find_galaxy(kd_tree, galaxy_pos_x, galaxy_pos_y, galaxy_type, galaxy_n_observed, galaxy_taken, center_x=0.0, center_y=0.0, radius=0.1):
    #find the galaxies that fall inside the observing radius
    index = kd_tree.query_ball_point([center_x,center_y], radius)
    selected_ID=-1
    
    # work only if there are galaxies within the patrol radius
    if (size(index) > 0):
        tmp_galaxy_ID = index
        tmp_galaxy_type = galaxy_type[index]
        tmp_n_observed = galaxy_n_observed[index]
        tmp_galaxy_taken  = galaxy_taken[index]
        
        id_QSO_I = where(tmp_galaxy_type==QSO_I)[0]
        id_QSO_II = where(tmp_galaxy_type==QSO_II)[0]
        id_LRG = where(tmp_galaxy_type==LRG)[0]
        id_ELG = where(tmp_galaxy_type==ELG)[0]

        # Look for QSO_II as a first priority
        if ((selected_ID==-1) & (id_QSO_II.size>0)):
            for i in id_QSO_II:
                if ((selected_ID==-1) &(tmp_n_observed[i]<5) & ( not tmp_galaxy_taken[i])):
                    selected_ID = tmp_galaxy_ID[i]
    
        # if nothing was found, then go for QSO_I
        if ((selected_ID==-1) & (id_QSO_I.size>0)):
            for i in id_QSO_I:
                if ((selected_ID==-1) &(tmp_n_observed[i]==0) & (not tmp_galaxy_taken[i])):
                    selected_ID = tmp_galaxy_ID[i]
        
        
        # if nothing was found, then go for LRGs
        if ((selected_ID==-1) & (id_LRG.size>0)):
            
            #priority is given to LRGs with one observation
            for i in id_LRG:
                if ((selected_ID==-1) & (tmp_n_observed[i]==1) & (not tmp_galaxy_taken[i])):
                    selected_ID = tmp_galaxy_ID[i]
                    
            for i in id_LRG: 
                if ((selected_ID==-1) & (tmp_n_observed[i]==0) & (not tmp_galaxy_taken[i])):
                    selected_ID = tmp_galaxy_ID[i]
                        
        
        # if nothing was found, then go for ELGs
        if ((selected_ID==-1) & (id_ELG.size>0)):
            for i in id_ELG:
                if ((selected_ID==-1) & (tmp_n_observed[i]==0) & (not tmp_galaxy_taken[i])):
                    selected_ID = tmp_galaxy_ID[i]
    
   # print "selected ID", selected_ID
    return selected_ID
```

```python
#compute kd-tree to speed-up the calculations
data = ones([n_total_gals,2])
data[:,0] = gal_pos_x[:]
data[:,1] = gal_pos_y[:]
data_tree = KDTree(data, leafsize=8)
```

```python
gal_ntimes_targeted[:] = 0

n_passes = 1
n_gal_type = 4
n_pointings = size(x_pointing)
n_fibers = size(positioner_x_deg)

print "n_pointings", n_pointings

# tracks the number of positioners used in each pointing, pass, per galaxy tipe
pointing_stats = zeros([n_pointings, n_passes, n_gal_type]) 


for i_pass in range(n_passes):
    for i_point in range(n_pointings):
        x_tile = x_pointing[i_point] + (2.0*patrol_radius * (random.random()-0.5))
        y_tile = y_pointing[i_point] + (2.0*patrol_radius * (random.random()-0.5))
        print i_pass, i_point
        gal_taken[:] = False
        for i_fiber in range(n_fibers):
            x_center = x_tile + positioner_x_deg[i_fiber]
            y_center = y_tile + positioner_y_deg[i_fiber]

            observed_ID = find_galaxy(data_tree, gal_pos_x, gal_pos_y, gal_type, gal_ntimes_targeted, gal_taken,
                                  center_x=x_center, center_y=y_center, radius=patrol_radius)
            
            if(observed_ID!=-1):
                gal_taken[observed_ID] = True
                gal_ntimes_targeted[observed_ID] = gal_ntimes_targeted[observed_ID]+1
                pointing_stats[i_point, i_pass, gal_type[observed_ID]]= pointing_stats[i_point, i_pass, gal_type[observed_ID]] + 1
            
                
```
